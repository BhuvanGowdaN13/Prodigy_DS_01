```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('population.csv')
data.head()
```

```python
data=data.drop(['Country Code','Indicator Name','Indicator Code'],axis=1)
data
```

```python
#Enter country name
def bargraph(country):
    cdata=data[data['Country Name']==country]

    x = cdata.columns[2:].astype(int)
    y = cdata.iloc[:,2:].values.flatten()

    plt.figure(figsize=(10,7))
    plt.bar(x, y, width=1, edgecolor="white", linewidth=0.7)
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.show()
```

```python
bargraph('India')
```

```python
bargraph('United States')
```

```python
bargraph('Aruba')
```
